```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl


data = pd.read_csv("/content/20200518_155835_JUN.csv")
data.drop(['Gear(dashboard)','Odometer','Avg_fuel_consumption','            Fuel_level'],axis=1,inplace=True)
data.drop(['            Coordinate(location)_x','Coordinate(location)_y','Coordinate(location)_z'],axis=1,inplace=True)
data.head(10)
```

```python
data2 = pd.read_csv("/content/20200518_160537_BANG.csv")
data2.drop(['Gear(dashboard)','Odometer','Avg_fuel_consumption','            Fuel_level'],axis=1,inplace=True)
data2.drop(['            Coordinate(location)_x','Coordinate(location)_y','Coordinate(location)_z'],axis=1,inplace=True)
data2.head(10)
```

```python
"""data.RPM.plot()
data2.RPM.plot()"""
plt.rcParams["figure.figsize"] = (15,10)
plt.plot(data.index,data.RPM,'g',label='JUN')
plt.plot(data2.index,data2.RPM,'b',label='BANG')
plt.title("RPM")
plt.xlabel("DELTA_TIME")
plt.ylabel("RPM")
plt.xlim(1000,6000)
plt.legend(loc='upper left')
plt.show()
"""data.Velocity.plot()
data2.Velocity.plot()"""
plt.plot(data.index,data.Velocity,'g',label='JUN')
plt.plot(data2.index,data2.Velocity,'b',label='BANG')
plt.title("VELOCITY")
plt.xlabel("DELTA_TIME")
plt.ylabel("VELOCITY(km/h)")
plt.xlim(1000,6000)
plt.legend(loc='upper left')
plt.show()
plt.plot(data.index,data.Accelerator,'g',label='JUN')
plt.plot(data2.index,data2.Accelerator,'b',label='BANG')
plt.title("ACCELERATOR")
plt.xlabel("DELTA_TIME")
plt.ylabel("VALUE")
plt.xlim(1000,6000)
plt.legend(loc='upper left')
plt.show()
plt.plot(data.index,data.Brake,'g',label='JUN')
plt.plot(data2.index,data2.Brake,'b',label='BANG')
plt.title("Brake")
plt.xlabel("DELTA_TIME")
plt.ylabel("VALUE")
plt.yscale("linear")
plt.xlim(1000,6000)
plt.legend(loc='upper left')
plt.show()
```
